=================================================================
## Task 2: Data Cleaning
=================================================================

```python
# STEP 1 — Convert Date Columns                      
date_columns = [
    "PQ First Sent to Client Date",
    "PO Sent to Vendor Date",
    "Scheduled Delivery Date",
    "Delivered to Client Date",
    "Delivery Recorded Date"
]

for column in date_columns:
    df[column] = pd.to_datetime(df[column], errors="coerce")
```

```python
# STEP 2 — Verify Date Conversion
df.info()
```

```python
# STEP 3 — Handle Missing Values 
df.isnull().sum()
```

```python
# Shipment Mode
df["Shipment Mode"] = df["Shipment Mode"].fillna(
    df["Shipment Mode"].mode()[0]
)
# Dosage
df["Dosage"] = df["Dosage"].fillna("Not Available")

# Insurance
df["Line Item Insurance (USD)"] = df[
    "Line Item Insurance (USD)"
].fillna(0)
```

```python
# STEP 4- Checking Missing Values Again
df.isnull().sum()
```

```python
# STEP 5 — Convert Numeric Columns
# Convert Weight
df["Weight (Kilograms)"] = pd.to_numeric(
    df["Weight (Kilograms)"],
    errors="coerce"
)

# Convert Freight Cost
df["Freight Cost (USD)"] = pd.to_numeric(
    df["Freight Cost (USD)"],
    errors="coerce"
)
```

```python
# STEP 6 — Verify Again
df.info()
```

```python
# STEP 7 — Save the Cleaned Dataset
df.to_csv("Cleaned_SupplyChain_Dataset.csv", index=False)
```

=================================================================
### Part B – Delivery Performance Analysis
###### Average delivery time
###### Delivery delay
###### Delayed shipments
=================================================================

```python
# Step 1: Calculate Delivery Delay
df["Delivery Delay"] = (
    df["Delivered to Client Date"]
    - df["Scheduled Delivery Date"]
).dt.days
```

```python
# Step 2: View the New Column
df[[
    "Scheduled Delivery Date",
    "Delivered to Client Date",
    "Delivery Delay"
]].head(10)
```

```python
print(df.columns)
```

```python
# Step 3: Average Delivery Delay
average_delay = df["Delivery Delay"].mean()

print("Average Delivery Delay:", average_delay)
```

Business Interpretation:
On average, shipments were delivered approximately 6 days earlier than the scheduled delivery date.

This suggests the supply chain is generally efficient and often completes deliveries ahead of schedule.

```python
# Step 4: Count Delayed Shipments
delayed_shipments = df[df["Delivery Delay"] > 0]

print("Total Delayed Shipments:")
print(delayed_shipments.shape[0])

# Delayed Shipment Percentage:
delay_percentage = (delayed_shipments.shape[0] / len(df)) * 100

print(f"Delayed Shipment Percentage: {delay_percentage:.2f}%")
```

Business Interpretation

Although 1,186 shipments experienced delays, the majority were delivered on time or ahead of schedule.

This indicates an overall reliable logistics network, while still highlighting an opportunity to investigate the causes of delayed deliveries.

## Task 4: Group-Based Analysis
========================================================

```python
# ==========================================
# Group Analysis 1
# Country vs Total Shipments
# ==========================================

country_shipments = (
    df.groupby("Country")
      .size()
      .sort_values(ascending=False)
)

print("Country vs Total Shipments")
print(country_shipments)
```

Business Interpretation

South Africa handled the highest number of shipments, followed by Nigeria and Côte d'Ivoire.

This indicates that these countries are the organization's primary distribution hubs and require stronger inventory management, transportation planning, and logistics coordination.

Countries with very few shipments (such as Belize and Guinea) likely represent smaller operational programs or lower customer demand.

```python
# ==========================================
# Group Analysis 2
# Vendor vs Total Cost
# ==========================================

vendor_cost = (
    df.groupby("Vendor")["Line Item Value"]
      .sum()
      .sort_values(ascending=False)
)

print("Vendor vs Total Cost")
print(vendor_cost.head(10))
```

Business Interpretation

One vendor (SCMS from RDC) accounts for a significantly larger procurement value than all others.

This suggests that the organization relies heavily on this supplier.

Although this may improve procurement efficiency, it also introduces supplier dependency risk. If that vendor experiences delays or shortages, the supply chain could be affected.

```python
# ==========================================
# Group Analysis 3
# Product Group vs Total Value
# ==========================================

product_value = (
    df.groupby("Product Group")["Line Item Value"]
      .sum()
      .sort_values(ascending=False)
)

print("Product Group vs Total Value")
print(product_value)
```

Business Interpretation

ARV (Antiretroviral medicines) contributes the vast majority of procurement value.

This indicates that HIV/AIDS treatment products are the primary focus of this supply chain.

Other product groups represent much smaller portions of the overall budget.

```python
# ==========================================
# Group Analysis 4
# Shipment Mode vs Average Delay
# ==========================================

shipment_delay = (
    df.groupby("Shipment Mode")["Delivery Delay"]
      .mean()
      .sort_values()
)

print("Shipment Mode vs Average Delivery Delay")
print(shipment_delay)
```

Business Interpretation

This is one of the strongest findings in your analysis.

Air Charter delivers shipments approximately 19 days earlier than scheduled on average.
Truck deliveries are also completed well ahead of schedule.
Standard Air transport performs efficiently with early deliveries.
Ocean shipments are the only mode with an average positive delay, meaning they tend to arrive about 6 days late.

This suggests that sea freight may require improved planning or should be reserved for non-urgent shipments.

```python
# Summary Tables
country_shipments.to_csv("Country_Shipments.csv")

vendor_cost.to_csv("Vendor_Total_Cost.csv")

product_value.to_csv("Product_Group_Value.csv")

shipment_delay.to_csv("Shipment_Mode_Delay.csv")
```

## Visualization 1: Country vs Total Shipments
========================================================

```python
import matplotlib.pyplot as plt

# Top 10 countries by shipment count
top10_country = df["Country"].value_counts().head(10)

plt.figure(figsize=(10,6))

plt.bar(top10_country.index,
        top10_country.values)

plt.title("Top 10 Countries by Total Shipments")

plt.xlabel("Country")

plt.ylabel("Number of Shipments")

plt.xticks(rotation=45)

plt.tight_layout()

# Save the chart
plt.savefig("Country_vs_Shipments.png",
            dpi=300,
            bbox_inches="tight")

# Display the chart

plt.show()
```

Business Interpretation:
The bar chart shows that South Africa received the highest number of shipments, followed by Nigeria and Côte d'Ivoire. These countries represent the organization's major distribution centers and require effective inventory management and logistics planning.

=======================================================
## Visualization 2: Shipment Mode Distribution
========================================================

```python
# Shipment Mode Distribution

shipment_mode = df["Shipment Mode"].value_counts()

plt.figure(figsize=(8,8))

plt.pie(
    shipment_mode.values,
    labels=shipment_mode.index,
    autopct="%1.1f%%",
    startangle=90
)

plt.title("Shipment Mode Distribution")

# Save the chart
plt.savefig("Shipment_Mode_Distribution.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

Business Interpretation:
The pie chart shows that Air transport accounts for the largest proportion of shipments, demonstrating that fast delivery is a priority within the supply chain. Truck transport is the second most frequently used mode, while Ocean shipments represent only a small percentage of total deliveries.

========================================================
## Visualization 3: Delivery Trend Over Time
========================================================

```python
# Delivery Trend Over Time

delivery_trend = (
    df.groupby(
        df["Delivered to Client Date"].dt.to_period("M")
    )
    .size()
)

plt.figure(figsize=(12,6))

plt.plot(
    delivery_trend.index.astype(str),
    delivery_trend.values,
    marker="o"
)

plt.title("Delivery Trend Over Time")

plt.xlabel("Month")

plt.ylabel("Number of Deliveries")

plt.xticks(rotation=45)

plt.tight_layout()

# Save the chart
plt.savefig("Delivery_Trend.png",
            dpi=300,
            bbox_inches="tight")

plt.show()
```

Business Interpretation:
The line chart illustrates the monthly delivery trend, allowing managers to identify periods of high and low shipment activity. Such trends can support demand forecasting, inventory planning, and logistics resource allocation.
